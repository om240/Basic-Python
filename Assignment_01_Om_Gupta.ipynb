## Basic string compression

```python
s='aaabbc'
res=''; count=1
for i in range(1, len(s)+1):
    if i<len(s) and s[i]==s[i-1]: count+=1
    else:
        res+=s[i-1]+str(count); count=1
res
```

## Remove duplicates (preserve order)

```python
lst=[1,2,2,3,4,4,5]
seen=set(); res=[]
for x in lst:
    if x not in seen: res.append(x); seen.add(x)
res
```

## Access nested dictionary safely

```python
d={'a':{'b':5}}
keys=['a','b']
val=d
for k in keys:
    val=val.get(k)
print(val)
```
